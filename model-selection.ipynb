```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score,KFold
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,StandardScaler,OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('gurgaon_properties_post_feature_selection_v2.csv')
```

```python
df.head()
```

```python
df['furnishing_type'].value_counts()
```

```python
# 0 is unfurnished,
#  1 is furnished, 
# 2 is semi-furnished
df['furnishing_type']=df['furnishing_type'].replace({0.0:'unfurnished',1.0:'furnished',2.0:'semi-furnished'})
```

```python
df.head(10)
```

```python
X=df.drop(columns=['price'])
y=df['price']
```

```python
# Split the data into training and testing sets
y_transformed=np.log(y)
```

Ordinal encoding


```python
columns_to_encode=['property_type','sector','balcony','agePossession','firnishing_type','luxury_category','floor_category']
```

```python
#create a column transformer for preprocessing
preprocessor=ColumnTransformer(
    transformers=[
        ('num',StandardScaler(),['area','bathrooms','bedrooms']),
        ('cat',OneHotEncoder(handle_unknown='ignore'),columns_to_encode)
    ],
    remainder='passthrough'
)
```

```python
#creating a pipeline for the model
pipeline=Pipeline([
    ('preprocessor',preprocessor),
    ('regressor',LinearRegression())
])
```

```python
# KFord cross validation
Kford=KFold(n_splits=5,shuffle=True,random_state=42)
scores=cross_val_score(pipeline,X,y_transformed,cv=Kford,scoring='r2')
```

```python
scores.mean(),scores.std()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y_transformed,test_size=0.2,random_state=42)
```

```python
Pipeline.fit(X_train,y_train)
```

```python
y_pred=pipeline.predict(X_test)
```

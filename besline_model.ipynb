```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('gurgaon_properties_post_feature_selection.csv')
df.head()
```

```python
x=df.drop(columns=['price'])
y=df['price']
```

```python
from sklearn.model_selection import KFold,cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
columns_to_encode=['sector','balcany','agePossession','furnishing','luxury_category','floor_category']
```

```python
#applying the log1p transformation to the target variable 
y_transformed=np.log1p(y)
```

```python
#creating the column transformer for preprocessing th
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['property_type','bedRoom','bathroom','built_up_area','servant room','store room']),
        ('cat', OneHotEncoder(drop='first'), columns_to_encode)],remainder='passthrough')
```

```python
#creating the pipeline

pipeline=Pipeline([
    ('preprocessor',preprocessor),
    ('regressor',LinearRegression())
])
```

```python
#K-Fold cross validation
from sklearn.model_selection import KFold, cross_val_score

kfold=KFold(n_splits=10,shuffle=True,random_state=42)
scores=cross_val_score(pipeline,x,y_transformed,cv=kfold,scoring='r2')
scores.mean()
```

```python
from sklearn.model_selection import KFold, cross_val_score
import pandas as pd

# Ensure x is a DataFrame and y_transformed is a Series or array
x = pd.DataFrame(x)
y_transformed = pd.Series(y_transformed)

# Reset index if needed
x = x.reset_index(drop=True)
y_transformed = y_transformed.reset_index(drop=True)

# Define K-Fold
kfold = KFold(n_splits=10, shuffle=True, random_state=42)

# Perform cross-validation
scores = cross_val_score(pipeline, x, y_transformed, cv=kfold, scoring='r2')
print(scores.mean())
```
